```python
import numpy as np
import pandas
from sklearn import datasets

print(np.__version__)
print(pandas.__version__ ) ##double underscore

## Start by pulling in a dataset?

digits = datasets.load_digits()

X = digits['data']   # this is the data with each 8x8 image "flattened" into a length-64 vector.
Y = digits['target'] # these are the labels (0-9). 
```

```python
## exploring the data set:

print("The shape of Y is:", Y.shape,"; it only has the 1 dimension")  ## see the dimensions?
print("The shape of X is:", X.shape,"; it has 2 dimensions")
print("There are ",X.shape[0],"; images in the data set")
print("and each image has ",X.shape[1]," pixels")
```

## What is happening?

X is an *object* in Python.  It has attributes that we can acces by adding a . and the name of the function. (This is an example of Object Oriented Programming (OOP).  

This object is a numpy.array, and we can look at the stuff inside of it with [square brackets].  
When I'm looking at a function (like shape) I call on its variables with (round brackets).  This is an important distinction!

```python
#If I want to look at a single record or row of X, I type:

n = 6 ## change me!

print(X[n])  #this is the nth X

#To see the corresponding Y, we type:
print(Y[n])
```

```python
##We've set this up so that the name is X, but the name could be anything:

num_set= X  ## I like to give things my name, because I'm not creative

num_set[n]
```

```python
## If I want to see just the mth number in X, (a "feature"), I use:
m = 12
X[n,m]

## note that counting starts at 0, so if m = 12, it is the 11th number in the row
#X[1,5]
```

```python
## what if I want to see a specific column?

print(X[:,m]) ## list of Column m

## I wonder if I can get the answer from the previous box?

print(X[:,m][n]) ##  Why :,?  It shows that i'm looking at the 2nd dimension

##or:

print(X[n][m])

```
